```python
count = 0
for i in range(len(data)):
    if data[i] == '(':
        count += 1
    elif data[i] == ')':
        count -= 1
    print(count,data[i])
print('Sum:', count)
```

```python
count = 0
for i in range(len(data)):
    if data[i] == '(':
        count += 1
    elif data[i] == ')':
        count -= 1
    print(i+1, count,data[i])
    if count == -1:
        break
print("Sum:", count, "step", i+1)
```
