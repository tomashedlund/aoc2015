```python
data = pd.read_csv("C:\\Users\\TOHED\\Desktop\\2015d5.txt", names=["content"])
```

```python
good = set()
for index, test in data.iterrows():
    if(len(test['content'])>3):
        for i in range(len(test['content'])-1):
            if test['content'].count(test['content'][i]+test['content'][i+1]) > 1:
                #print(test['content'], test['content'][i]+test['content'][i+1])
                good.add(index)
                break
            
data = data.loc[good]
len(data)
```
