```python
f = open("C:\\Users\\TOHED\\Desktop\\2015d3.txt", "r")
data = f.read()
f.close()
```

```python
x = 0
y = 0
coordinates = set([(0, 0)])
coordinates
```

```python
for c in data:
    if c == '^':
        y += 1
    elif c == 'v':
        y -= 1
    elif c == '>':
        x += 1
    elif c == '<':
        x -= 1
    coordinates.add((x,y))
print(len(coordinates))
```

```python
x = 0
y = 0
x2 = 0
y2 = 0
santa = True
coordinates = set([(0,0)])
coordinates
```

```python
for c in data:
    if santa:
        if c == '^':
            y += 1
        elif c == 'v':
            y -= 1
        elif c == '>':
            x += 1
        elif c == '<':
            x -= 1
        santa = False
        coordinates.add((x,y))
    else:
        if c == '^':
            y2+= 1
        elif c == 'v':
            y2-= 1
        elif c == '>':
            x2+= 1
        elif c == '<':
            x2-= 1
        santa = True
        coordinates.add((x2,y2))
print(len(coordinates))
```
